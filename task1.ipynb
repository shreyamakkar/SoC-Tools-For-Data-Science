```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
plt.rcParams['figure.figsize'] = (12.0, 9.0)

data = pd.read_csv('task1.csv')
X = data.iloc[:, 0]
Y = data.iloc[:, 1]
plt.scatter(X, Y)
plt.xlabel('No. of days')
plt.ylabel('Total cases')
plt.show()

m=0
c=0
m1=0
c1=0 #y = mx+c where m is c2, c is logc1, x=number of days, y= log(total number of cases)
L = 0.0001 #learning rate
L1 = 1
iterations = 5000 #number of iterations
n = float(len(X)) #number of elements in X

for i in range(iterations):
    Y_pred = m*X + c 
    D_m = (-2/n)*sum(X*(np.log(Y)-Y_pred))
    D_c = (-2/n)*sum(np.log(Y)-Y_pred)
    m = m - L*D_m
    c = c - L*D_c


for i in range(iterations):
    Y_pred = m1*X + c 
    D_m1 = (-2/n)*sum(X*(np.log(Y)-Y_pred))
    D_c1 = (-2/n)*sum(np.log(Y)-Y_pred)
    m1 = m1 - L1*D_m1
    c1 = c1 - L1*D_c1


print(m, c)
plt.scatter(X, np.log(Y))
plt.plot(X, m*X + c)
plt.title('Linear fit')
plt.xlabel('No. of days')
plt.ylabel('Log of total cases')
plt.legend([r'$\alpha=0.0001$'])

if (np.exp(m*110 + c)) < 100000:
    print(np.exp(m*110 + c))
    print("Yes, the hospitals can handle the cases")
else: 
    print(np.exp(m*110 + c))
    print("No, the hospitals can't handle the cases")

plt.show()
''' If the learning rate is 1, the model learns faster and diverges from the given graph, thus giving wrong output. If the learning rate is 0.00001, it will take the model more time in learning and will have smaller updates in the values of m and c.'''

```
